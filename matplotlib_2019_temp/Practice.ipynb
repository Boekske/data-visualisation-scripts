```python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib.patches import Circle
from matplotlib.offsetbox import (TextArea, DrawingArea, OffsetImage, AnnotationBbox)
from matplotlib.cbook import get_sample_data
```

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.axis('equal')
# Annotate the 2nd position with a circle patch

ax.add_artist(Circle((1, .5), .15, transform=fig.transFigure))

# Fix the display limits to see everything
ax.set_xlim(0, 2)
ax.set_ylim(0, 2)

plt.show()
```

```python
# Set canvas background color the same as axes
plt.rc('figure', facecolor='w')

# Arrange city based on latitude
cityList = ['hobart', 'melbourne', 'canberra', 'adelaide', 'sydney', 'perth', 'brisbane', 'darwin']

# set constrianed_layout as True to avoid axes overlap
fig = plt.figure(figsize=(10,12), dpi=50, constrained_layout=True)
# Use GridSpec for customising laypout
gs = fig.add_gridspec(nrows=10, ncols=1)

# Set up a empty axes that occupied 2 rows for text
axText = fig.add_subplot(gs[0:2, :])

# Add a color bar 
axInlet = fig.add_axes([0.1, 0.86, 0.08, 0.01], frameon=True)

for i, city in enumerate(cityList):
    ax = fig.add_subplot(gs[i+2, :], ylim=(-30, 30))
    
    ax.set_ylabel(city.title(), rotation=0, color='k', fontsize=10, fontweight='bold')
    
plt.show()
```
