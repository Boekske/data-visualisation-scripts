# `figure` with multiple `axes` 

### Subplots with regular layout

```python
fig = plt.figure()
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)

ax1.annotate('I am ax1', (0.5, 0.5),
             xycoords='axes fraction', va='center', ha='center')
ax2.annotate('I am ax2', (0.5, 0.5),
             xycoords='axes fraction', va='center', ha='center')
ax3.annotate('I am ax3', (0.5, 0.5),
             xycoords='axes fraction', va='center', ha='center')
ax4.annotate('I am ax4', (0.5, 0.5),
             xycoords='axes fraction', va='center', ha='center')

plt.tight_layout()
```

#### Access items by flattening the ndarray

```python
fig = plt.figure()
axs = fig.subplots(nrows=2, ncols=2)

axs.flat[0].annotate('I am ax1', (0.5, 0.5),
               xycoords='axes fraction', va='center', ha='center')
axs.flat[1].annotate('I am ax2', (0.5, 0.5),
               xycoords='axes fraction', va='center', ha='center')
axs.flat[2].annotate('I am ax3', (0.5, 0.5),
               xycoords='axes fraction', va='center', ha='center')
axs.flat[3].annotate('I am ax4', (0.5, 0.5),
               xycoords='axes fraction', va='center', ha='center')

plt.tight_layout()
```

#### Tuple unpacking

```python
fig = plt.figure()
((ax1, ax2), (ax3, ax4)) = fig.subplots(nrows=2, ncols=2)

ax1.annotate('I am ax1', (0.5, 0.5),
               xycoords='axes fraction', va='center', ha='center')
ax2.annotate('I am ax2', (0.5, 0.5),
               xycoords='axes fraction', va='center', ha='center')
ax3.annotate('I am ax3', (0.5, 0.5),
               xycoords='axes fraction', va='center', ha='center')
ax4.annotate('I am ax4', (0.5, 0.5),
               xycoords='axes fraction', va='center', ha='center')

plt.tight_layout()
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0, 0.6, 0.5, 0.5])
ax2 = fig.add_axes([0.6, 0.6, 0.5, 0.5])
ax3 = fig.add_axes([0, 0, 0.5, 0.5])
ax4 = fig.add_axes([0.6, 0, 0.5, 0.5])

ax1.annotate('I am ax1', (0.5, 0.5),
               xycoords='axes fraction', va='center', ha='center')
ax2.annotate('I am ax2', (0.5, 0.5),
               xycoords='axes fraction', va='center', ha='center')
ax3.annotate('I am ax3', (0.5, 0.5),
               xycoords='axes fraction', va='center', ha='center')
ax4.annotate('I am ax4', (0.5, 0.5),
               xycoords='axes fraction', va='center', ha='center')

plt.show()
```

### Subplots with unregular layout

`pyplot.subplots` creates a figure and a grid of subplots with a single call, while providing reasonable control over how the individual plots are created. For more advanced use cases you can use `GridSpec` for a more general subplot layout or `Figure.add_subplot` for adding subplots at arbitrary locations within the figure.

```python
fig = plt.figure()
ax1 = plt.subplot(221)
ax2 = plt.subplot(223)
ax3 = plt.subplot(122)
plt.tight_layout()
```

```python
fig = plt.figure()
ax1 = plt.subplot(10, 10, 3) 
ax2 = plt.subplot(223) # Note that all integers must be less than 10 for this form to work
ax3 = plt.subplot(122)
plt.show()
```

### what if it is a array?

```python
fig, axs = plt.subplots(nrows=2, ncols=2, constrained_layout=True)
axs.flat[0].annotate('I am ax0', (0.5, 0.5),
               xycoords='axes fraction', va='center', ha='center')
axs.flat[1].annotate('I am ax1', (0.5, 0.5),
               xycoords='axes fraction', va='center', ha='center')
axs.flat[2].annotate('I am ax2', (0.5, 0.5),
               xycoords='axes fraction', va='center', ha='center')
axs.flat[3].annotate('I am ax3', (0.5, 0.5),
               xycoords='axes fraction', va='center', ha='center')
```

```python
fig, axs = plt.subplots(nrows=2, ncols=2, constrained_layout=True)
axs[0, 0].annotate('I am ax0', (0.5, 0.5),
               xycoords='axes fraction', va='center', ha='center')
axs[0, 1].annotate('I am ax1', (0.5, 0.5),
               xycoords='axes fraction', va='center', ha='center')
axs[1, 0].annotate('I am ax2', (0.5, 0.5),
               xycoords='axes fraction', va='center', ha='center')
axs[1, 1].annotate('I am ax3', (0.5, 0.5),
               xycoords='axes fraction', va='center', ha='center')
```

```python
# You can use tuple-unpacking also in 2D to assign all subplots to dedicated variables
fig, ((ax0, ax1), (ax2, ax3)) = plt.subplots(nrows=2, ncols=2, constrained_layout=True)
ax0.annotate('I am ax0', (0.5, 0.5),
               xycoords='axes fraction', va='center', ha='center')
ax1.annotate('I am ax1', (0.5, 0.5),
               xycoords='axes fraction', va='center', ha='center')
ax2.annotate('I am ax2', (0.5, 0.5),
               xycoords='axes fraction', va='center', ha='center')
ax3.annotate('I am ax3', (0.5, 0.5),
               xycoords='axes fraction', va='center', ha='center')
```

### Subplot share y-axis

```python
fig, (ax0, ax1) = plt.subplots(ncols=2, constrained_layout=True)
```

```python
fig, (ax0, ax1) = plt.subplots(ncols=2, sharey=True, constrained_layout=True)
```

```python
fig = plt.figure()
ax1 = plt.subplot(221)
ax2 = ax1.twinx()
ax3 = plt.subplot(122)
plt.tight_layout()
```

```python
fig = plt.figure()
ax1 = plt.axes([0, 0.6, 0.5, 0.5])
ax2 = plt.axes([0, 0, 0.5, 0.5])
ax3 = plt.axes([0.6, 0, 0.5, 1.1])
```

```python
fig, axs = plt.subplots(ncols=3, nrows=3)
gs = axs[1, 2].get_gridspec()
# remove the underlying axes
for ax in axs[1:, -1]:
    ax.remove()
axbig = fig.add_subplot(gs[1:, -1])
axbig.annotate('Big Axes \nGridSpec[1:, -1]', (0.1, 0.5),
               xycoords='axes fraction', va='center')

fig.tight_layout()

plt.show()
```

```python
# Share a X axis with each column of subplots
fig = plt.figure()
axs = plt.subplots(2, 2, sharex='col')
```

```python
# Share both X and Y axes with all subplots
plt.subplots(2, 2, sharex='all', sharey='all')
plt.show()
```

```python
# Creates figure number 10 with a single subplot
# and clear it if it already exists.
fig, ax=plt.subplots(num=10, clear=True)
```
