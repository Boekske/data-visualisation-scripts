```python
import glob
import pandas as pd
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

import numpy as np
import seaborn as sns
from datetime import datetime

import matplotlib as mpl
from matplotlib import cm
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker 
import matplotlib.animation as animation
from IPython.display import HTML

# enable matplotlib display static plot within jupyter notebook
%matplotlib inline 
```

```python
# set up plot style
from matplotlib import cycler
colors = cycler('color',
                ['#8dd3c7', '#ffffb3', '#bebada', '#fb8072', '#80b1d3', '#fdb462', '#b3de69'])
plt.rc('figure', facecolor='#222222')
plt.rc('axes', facecolor='#222222', edgecolor='#222222',
       axisbelow=True, grid=True, prop_cycle=colors) # "axisbelow" set axis ticks and gridlines are below all artists
plt.rc('grid', color='#e3e3e3', linestyle='solid')
plt.rc('xtick', direction='in', color='#e3e3e3')
plt.rc('ytick', direction='in', color='#e3e3e3')
plt.rc('patch', edgecolor='#e3e3e3')
plt.rc('lines', linewidth=1.5)
```

```python
df_sydney=pd.read_csv('./main-city-temp/sydney-max-tmp.csv')
df_sydney.head()
```

```python
df_sydney['date'] = df_sydney['Year'].astype(str) + '-' + df_sydney['Month'].astype(str) + '-' + df_sydney['Day'].astype(str)
df_sydney.head()
```

```python
df_sydney['date']=[datetime.strptime(d, '%Y-%m-%d').date() for d in df_sydney['date']]
df_sydney.head()
```

```python
df_sydney.rename(columns={'Maximum temperature (Degree C)':'max_tmp_day'}, inplace=True)
df_sydney.head()
```

```python
df_sydney['max_tmp_year']=df_sydney.groupby('Year')['max_tmp_day'].transform('max')
df_sydney.head()
```

```python
df_sydney[['Days of accumulation of maximum temperature']]=df_sydney[['Days of accumulation of maximum temperature']].fillna(0)
df_sydney.head()
```

```python
df_sydney['diff']=df_sydney['max_tmp_year'] - df_sydney['max_tmp_day']
df_sydney.head()
```

```python
df_sydney['hot'] = np.NaN
df_sydney.head()
```

```python
for i in range(df_sydney.shape[0]):
    if df_sydney['diff'][i] <= (df_sydney['max_tmp_year'][i] - df_sydney['max_tmp_year'][i]*0.95):
        df_sydney['hot'][i] = 1

df_sydney.head()
```

```python
df_sydney['hot_day']=df_sydney.groupby('Year')['hot'].transform('sum')
df_sydney.head()
```

```python
fig, ax1 = plt.subplots(figsize=(12,4), dpi=200)
ax2 = ax1.twinx()

ax1.clear()
ax2.clear()

ax1.plot(df_sydney.Year, df_sydney.max_tmp_year*0, color='#222222', zorder=1)


ax1.plot(df_sydney.Year, df_sydney.max_tmp_year, color='#fc280f', zorder=2)
ax2.plot(df_sydney.Year, df_sydney.hot_day, color='#eeb72b', zorder=2)

ax1.scatter(df_sydney.Year.values[-1], df_sydney.max_tmp_year.values[-1], color='w', s=100, edgecolor ='k', linewidth=1, zorder=3)
ax2.scatter(df_sydney.Year.values[-1], df_sydney.hot_day.values[-1], color='w', s=100, edgecolor ='k', linewidth=1, zorder=3)

ax1.text(df_sydney.Year.values[-1], df_sydney.max_tmp_year.values[-1]+5, 
         s='Temp. {}'.format(df_sydney.max_tmp_year.values[-1]), size=10, color='#e3e3e3', ha='center', va='bottom', zorder=6)
ax2.text(df_sydney.Year.values[-1], df_sydney.hot_day.values[-1]+1, 
         s='{} Days'.format(df_sydney.hot_day.values[-1]), size=10, color='#e3e3e3', ha='center', va='top', zorder=6)

ax1.text(0.5, 0.5, s=df_sydney.Year.values[-1], fontweight='bold', size=60, color='#e3e3e3', ha='center', va="center", alpha=0.5, transform=ax1.transAxes, zorder=1)

#ax.legend()
ax1.set_ylim(0, df_sydney.max_tmp_year.max()*1.68)
ax1.grid(b=None)
ax2.grid(b=None) # Don't show grid of ax2 
ax1.set_ylabel('Temperature (oC)', color='#e3e3e3')
ax2.set_ylabel('Day', color='#e3e3e3')
ax2.spines['bottom'].set_color('#e3e3e3')  
ax2.spines['right'].set_color('#e3e3e3')  
ax2.spines['left'].set_color('#e3e3e3')  
ax2.spines['top'].set_color('#e3e3e3') 

# Retrieve an element of a plot and set properties
for tick in ax1.xaxis.get_ticklabels():
    tick.set_fontsize('x-small')
    tick.set_rotation(0)

ax1.margins(x=0.05)

ax1.text(0.00, 1.05, 'Sydney Yearly Maximum Temperature From 1859 to 2020', color='#e3e3e3', transform=ax1.transAxes, size=17, weight='light', ha='left')

fig.canvas.draw()
```

```python
df_sydney=pd.read_csv('./main-city-temp/sydney-max-tmp.csv')

def data_clean(df):
    df['date'] = df['Year'].astype(str) + '-' + df['Month'].astype(str) + '-' + df['Day'].astype(str)
    df['date']=[datetime.strptime(d, '%Y-%m-%d').date() for d in df['date']]
    df.rename(columns={'Maximum temperature (Degree C)':'max_tmp_day'}, inplace=True)
    df['max_tmp_year']=df.groupby('Year')['max_tmp_day'].transform('max')
    return df

df_sydney = data_clean(df_sydney)
df_sydney.head()
```

```python
from os.path import basename

all_files = glob.glob("./main-city-temp/sydney-max-tmp.csv")

d={}
for filename in all_files:
    d[basename(filename)[:-12]]=pd.read_csv(filename)
    d[basename(filename)[:-12]]=data_clean(d[basename(filename)[:-12]])
```

```python
d['sydney'].shape
```

```python
fig, ax = plt.subplots(figsize=(12,4), dpi=300)

ax.clear()
ax.plot(d['sydney'].Year, d['sydney'].max_tmp_year*0, color='#E6E6E6', zorder=1)

for city in d.keys():
    ax.plot(d[city].Year, d[city].max_tmp_year, zorder=2, alpha=0.8)
    ax.scatter(d[city].Year.values[-1], d[city].max_tmp_year.values[-1], 
               color='w', s=100, edgecolor ='k', linewidth=1, zorder=3)
    ax.text(d[city].Year.values[-1], d[city].max_tmp_year.values[-1]+1, 
            s='{} {}'.format(city.title(), d[city].max_tmp_year.values[-1]), size=5, ha='center', va='bottom', zorder=6)
    
ax.text(0.5, 0.5, s=d['sydney'].Year.values[-1], fontweight='bold', size=60, ha='center', va="center", alpha=0.5, transform=ax.transAxes, zorder=1)

ax.set_ylim(20, d['sydney'].max_tmp_year.max()*1.3)

# Retrieve an element of a plot and set properties
for tick in ax.xaxis.get_ticklabels():
    tick.set_fontsize('x-small')
    tick.set_rotation(0)

ax.margins(x=0.01)

fig.canvas.draw()
```

```python
def draw_areachart(Num_Date):

    ax.clear()
        
    df_temp=df.loc[0:Num_Date,:]
    
        
    #### Draw artists in axes
    ax.plot(df.date, df.price*0, color='#E6E6E6', zorder=1) # plot a axes with only date as x-axis and no value for y-axis, set its color the same as bg

    ax.plot(df_temp.date, df_temp.price, label='BTC', zorder=2) # line plot for first group
    ax.plot(df_temp.date, df_temp.price*10, label='ETC', zorder=3) # line plot for second group

    ax.scatter(df_temp.date.values[-1], df_temp.price.values[-1], color='w',s=100,edgecolor ='k',linewidth=1, zorder=4)
    ax.scatter(df_temp.date.values[-1], df_temp.price.values[-1]*10, color='w',s=100,edgecolor ='k',linewidth=1, zorder=5)

    ax.text(df_temp.date.values[-1], df_temp.price.values[-1]*1.28, s='BTC {}'.format(np.round(df_temp.price.values[-1],1)), size=10, ha='left', va='bottom', zorder=6)
    ax.text(df_temp.date.values[-1], df_temp.price.values[-1]*1.28*10, s='ETC {}'.format(np.round(df_temp.price.values[-1]*10,1)), size=10, ha='left', va='bottom', zorder=7)
    
    # transform indicates coordinate text position relative to axes not real value
    ax.text(0.5, 0.5, s=df_temp.date.values[-1].year, fontweight='bold', size=60, ha='center', va="center", alpha=0.5, transform=ax.transAxes, zorder=1) 
    #ax.legend()
    
    #### adjust axes features
    ax.set_ylim(0, df_temp.price.max()*1.68*10)
    ax.set_xticks(ticks=df.date.values[::365])

    # Retrieve an element of a plot and set properties
    for tick in ax.xaxis.get_ticklabels():
        tick.set_fontsize('x-small')
        tick.set_rotation(-75) # set rotation angle 
    
    #### set axes position and text
    ax.margins(x=0.01)
    ax.set_ylabel("Currency average price ($)", size=10)
    ax.text(-0.07, 1.03, 'BTC price change between 2013 to 2019',transform=ax.transAxes, size=17, weight='light', ha='left')
```
